```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_test = pd.read_csv('Raw datasets/war_daily_bat.csv')
df_batting = pd.read_csv('Raw datasets/batting.csv')
df_fielding = pd.read_csv('Raw datasets/fielding.csv')
df_people = pd.read_csv('Raw datasets/people.csv')
df_teams = pd.read_csv('Raw datasets/teams.csv')
df_salaries = pd.read_csv('Raw datasets/Salaries.csv')
```

```python
# mean WAR over the years
df_test.groupby('year_ID')['WAR'].mean().plot(kind='bar', figsize=(15, 10))
```

```python
# PA over the years
df_test.groupby('year_ID')['G'].sum().plot(kind='line', figsize=(15, 10))
```

```python
# Mean PA over the years
df_test.groupby('year_ID')['PA'].sum().plot(kind='line', figsize=(15, 10))
```

```python
# Join df_test and df_batting on player_ID = playerID and year_ID = yearID
df_test_batting = pd.merge(df_test, df_batting, how='inner', left_on=['player_ID', 'year_ID'], right_on=['playerID', 'yearID'])
```

```python
# aaron judge
df_test_batting[df_test_batting['name_common'] == 'Aaron Judge']
```

```python
# Join df_test_batting and df_people on player_ID = playerID
df_test_batting_people = pd.merge(df_test_batting, df_people, how='inner', left_on=['player_ID'], right_on=['playerID'])
```

```python
# Derek Jeter
df_test_batting_people[df_test_batting_people['name_common'] == 'Derek Jeter']
```

```python
# Join df_test_batting_people and df_teams on team_ID = teamID and year_ID = yearID
df_test_batting_people_teams = pd.merge(df_test_batting_people, df_teams, how='inner', left_on=['team_ID', 'year_ID'], right_on=['franchID', 'yearID'])
```

```python
# Derek Jeter
df_test_batting_people_teams[df_test_batting_people_teams['name_common'] == 'Derek Jeter'].sort_values(by='year_ID')
```

```python
# Join salaries and df_test_batting_people_teams on player_ID = playerID and year_ID = yearID
df_test_batting_people_teams_salaries = pd.merge(df_test_batting_people_teams, df_salaries, how='inner', left_on=['player_ID', 'year_ID'], right_on=['playerID', 'yearID'])
```
