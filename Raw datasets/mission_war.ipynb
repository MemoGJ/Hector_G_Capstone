```python
import pandas as pd
import numpy as np
```

```python
# Load data
df = pd.read_csv('war_daily_bat.csv')
```

```python
# Columns to keep
cols_to_keep = ['name_common',
 'age',
 'mlb_ID',
 'player_ID',
 'year_ID',
 'team_ID',
 'stint_ID',
 'lg_ID',
 'PA',
 'G',
 'Inn', 'salary', 'OPS_plus', 'WAR', 'WAR_def', 'WAR_off', 'pitcher']
```

```python
# Drop columns
df = df[cols_to_keep]
```

```python
df.sample(5)
```

```python
# Joey Gallo and Steve Pearce
joey_pearce = df[(df['name_common'] == 'Joey Gallo') | (df['name_common'] == 'Steve Pearce')]
joey_pearce
```

```python
# New dataframe for joey_pearce with player_ID, year_ID, and Team_ID only
joey_pearce_teams = joey_pearce[['player_ID', 'year_ID', 'team_ID']]
joey_pearce_teams
```

```python
# Group by player_ID and year_ID, keep only the first team_ID
joey_pearce_teams = joey_pearce_teams.groupby(['player_ID', 'year_ID']).first()
joey_pearce_teams
```

```python
# group by name_common, age, year_ID, keep player_ID, team_ID
joey_pearce_group = joey_pearce.groupby(['name_common', 'age', 'year_ID', 'player_ID', 'pitcher']).sum().reset_index()
joey_pearce_group
```

```python
# Join joey_pearce_group and joey_pearce_teams on player_ID and year_ID
joey_pearce_join = joey_pearce_group.merge(joey_pearce_teams, on=['player_ID', 'year_ID'])
joey_pearce_join
```

```python
# New dataframe for df with player_ID, year_ID, and Team_ID only
df_teams = df[['player_ID', 'year_ID', 'team_ID', 'stint_ID']]
df_teams.shape
```

```python
df_teams.head()
```

```python
# drop rows with stint_ID > 1
df_teams = df_teams[df_teams['stint_ID'] == 1]
```

```python
# drop stint_ID column
df_teams = df_teams.drop('stint_ID', axis=1)
```

```python
# group by name_common, age, year_ID, keep player_ID, team_ID, df
df_group = df.groupby(['name_common', 'age', 'year_ID', 'player_ID', 'pitcher']).sum().reset_index()
```

```python
df_group.shape
```

```python
df_group.sample(10)
```

```python
# Join df_group and df_teams on player_ID and year_ID
df_join = df_group.merge(df_teams, on=['player_ID', 'year_ID'])
```
