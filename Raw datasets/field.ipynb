```python
# subset from 1985
df = df[df["yearID"] >= 1985]
```

```python
# Unique players
df["playerID"].nunique()
```

```python
# Drop catcher columns
df = df.drop(["PB", "WP", "SB", "CS", 'ZR'], axis=1)
```

```python
# Group by playerID and POS
positions = df.groupby(["playerID", "POS"])["G"].sum().reset_index().sort_values("G", ascending=False).groupby("playerID").first()

# Index to column
positions = positions.reset_index()

# drop G column
positions = positions.drop("G", axis=1)

```
